```python
# 1. ÿ™ŸÜÿ∏ŸäŸÅ ÿßŸÑÿ®Ÿäÿ¶ÿ© ÿ™ŸÖÿßŸÖÿßŸã ŸÑŸÑÿ®ÿØÿ° ŸÖŸÜ ÿ¨ÿØŸäÿØ
!rm -rf .buildozer
!rm -rf bin
!rm -rf main.py buildozer.spec

# 2. ÿ™ÿ´ÿ®Ÿäÿ™ ÿßŸÑŸÖŸÉÿ™ÿ®ÿßÿ™ ÿßŸÑÿ£ÿ≥ÿßÿ≥Ÿäÿ© (ÿ®ŸÖÿß ŸÅŸäŸáÿß ÿßŸÑÿ≠ŸÑ ŸÑŸÖÿ¥ŸÉŸÑÿ© libtool)
!sudo apt-get update
!sudo apt-get install -y build-essential libtool libtool-bin autoconf automake \
  python3-pip python3-dev ffmpeg libsdl2-dev libsdl2-image-dev libsdl2-mixer-dev \
  libsdl2-ttf-dev libportmidi-dev libswscale-dev libavformat-dev libavcodec-dev \
  zlib1g-dev libgstreamer1.0-0 gstreamer1.0-plugins-base gstreamer1.0-plugins-good \
  libssl-dev libffi-dev zip unzip

# 3. ÿ™ÿ´ÿ®Ÿäÿ™ Buildozer Ÿà Cython (ŸÜÿ≥ÿÆÿ© ŸÖÿ™ŸàÿßŸÅŸÇÿ©)
!pip install buildozer cython==0.29.33

# 4. ŸÉÿ™ÿßÿ®ÿ© ŸÉŸàÿØ ÿßŸÑÿ™ÿ∑ÿ®ŸäŸÇ (main.py)
code = """
from kivymd.app import MDApp
from kivymd.uix.screen import MDScreen
from kivymd.uix.textfield import MDTextField
from kivymd.uix.button import MDFillRoundFlatButton
from kivymd.uix.list import MDList, TwoLineListItem, OneLineListItem
from kivymd.uix.scrollview import MDScrollView
from kivymd.uix.boxlayout import MDBoxLayout
from kivy.clock import Clock
from kivy.utils import platform
import yt_dlp
import threading
import os

# ÿµŸÑÿßÿ≠Ÿäÿßÿ™ ŸàŸÖÿ≥ÿßÿ± ÿßŸÑÿ≠ŸÅÿ∏
if platform == 'android':
    from android.permissions import request_permissions, Permission
    request_permissions([Permission.READ_EXTERNAL_STORAGE, Permission.WRITE_EXTERNAL_STORAGE])
    download_path = "/storage/emulated/0/Download/MyTube"
else:
    download_path = "Download_Test"

if not os.path.exists(download_path):
    try:
        os.makedirs(download_path)
    except:
        pass

class MyTubeApp(MDApp):
    def build(self):
        self.theme_cls.primary_palette = "Red"
        self.theme_cls.theme_style = "Dark"

        screen = MDScreen()
        layout = MDBoxLayout(orientation='vertical', padding=20, spacing=20)

        self.url_input = MDTextField(hint_text="ÿ±ÿßÿ®ÿ∑ ÿßŸÑŸÅŸäÿØŸäŸà (Link)", mode="rectangle")
        btn_search = MDFillRoundFlatButton(text="ÿ®ÿ≠ÿ´ (Search)", pos_hint={'center_x': 0.5}, on_release=self.start_search)

        scroll = MDScrollView()
        self.list_view = MDList()
        scroll.add_widget(self.list_view)

        layout.add_widget(self.url_input)
        layout.add_widget(btn_search)
        layout.add_widget(scroll)
        screen.add_widget(layout)
        return screen

    def start_search(self, instance):
        url = self.url_input.text
        if not url: return
        self.list_view.clear_widgets()
        self.list_view.add_widget(OneLineListItem(text="ÿ¨ÿßÿ±Ÿä ÿßŸÑÿ®ÿ≠ÿ´..."))
        threading.Thread(target=self.fetch_info, args=(url,)).start()

    def fetch_info(self, url):
        try:
            ydl_opts = {'quiet': True, 'no_warnings': True}
            with yt_dlp.YoutubeDL(ydl_opts) as ydl:
                info = ydl.extract_info(url, download=False)
                Clock.schedule_once(lambda x: self.update_list(info, url))
        except Exception as e:
            Clock.schedule_once(lambda x: self.show_error(str(e)))

    def update_list(self, info, url):
        self.list_view.clear_widgets()
        self.list_view.add_widget(OneLineListItem(text=info.get('title', '')[:40]))

        # ÿÆŸäÿßÿ±ÿßÿ™ ÿßŸÑÿµŸàÿ™
        self.list_view.add_widget(TwoLineListItem(text="ÿ™ÿ≠ŸÖŸäŸÑ ÿµŸàÿ™ (MP3/Audio)", secondary_text="Best Quality", on_release=lambda x: self.download_video(url, 'audio')))

        # ÿÆŸäÿßÿ±ÿßÿ™ ÿßŸÑŸÅŸäÿØŸäŸà
        formats = info.get('formats', [])
        seen = set()
        # ÿ™ÿ±ÿ™Ÿäÿ® ÿßŸÑÿ¨ŸàÿØÿßÿ™ ŸÖŸÜ ÿßŸÑÿ£ŸÅÿ∂ŸÑ ŸÑŸÑÿ£ÿ≥Ÿàÿ£
        formats.sort(key=lambda x: x.get('height') or 0, reverse=True)

        for f in formats:
            h = f.get('height')
            ext = f.get('ext')
            if h and h >= 144 and ext in ['mp4', 'webm']:
                res_str = f"{h}p"
                if res_str not in seen:
                    seen.add(res_str)
                    self.list_view.add_widget(TwoLineListItem(text=f"ŸÅŸäÿØŸäŸà {res_str}", secondary_text=f"Type: {ext}", on_release=lambda x, fid=f['format_id']: self.download_video(url, 'video', fid)))

    def download_video(self, url, type, fid=None):
        self.list_view.clear_widgets()
        self.list_view.add_widget(OneLineListItem(text="ÿ¨ÿßÿ±Ÿä ÿßŸÑÿ™ÿ≠ŸÖŸäŸÑ... ÿßŸÜÿ™ÿ∏ÿ±"))
        threading.Thread(target=self.run_download, args=(url, type, fid)).start()

    def run_download(self, url, type, fid):
        try:
            # ŸáŸÜÿß ŸÜÿ≥ÿ™ÿÆÿØŸÖ bestvideo+bestaudio ŸàŸÑŸÉŸÜ ŸÇÿØ Ÿäÿ™ÿ∑ŸÑÿ® ffmpeg ŸÖÿØŸÖÿ¨
            # ŸÑŸÑÿ™ÿ®ÿ≥Ÿäÿ∑ ÿ≥ŸÜÿ≠ŸÖŸÑ best ÿßŸÑŸÖÿ™ÿßÿ≠
            opts = {
                'outtmpl': f'{download_path}/%(title)s.%(ext)s',
                'quiet': True,
                'ignoreerrors': True
            }

            if type == 'audio':
                opts['format'] = 'bestaudio/best'
            else:
                if fid:
                    opts['format'] = f"{fid}+bestaudio/best"
                else:
                    opts['format'] = "best"

            with yt_dlp.YoutubeDL(opts) as ydl:
                ydl.download([url])
            Clock.schedule_once(lambda x: self.show_success())
        except Exception as e:
            Clock.schedule_once(lambda x: self.show_error(str(e)))

    def show_success(self):
        self.list_view.clear_widgets()
        self.list_view.add_widget(OneLineListItem(text="‚úÖ ÿ™ŸÖ ÿßŸÑÿ™ÿ≠ŸÖŸäŸÑ ÿ®ŸÜÿ¨ÿßÿ≠!"))
        self.list_view.add_widget(OneLineListItem(text=f"ÿßŸÑŸÖŸÑŸÅ ŸÅŸä: {download_path}"))

    def show_error(self, err):
        self.list_view.clear_widgets()
        self.list_view.add_widget(OneLineListItem(text=f"ÿÆÿ∑ÿ£: {err}"))

if __name__ == '__main__':
    MyTubeApp().run()
"""
with open("main.py", "w") as f:
    f.write(code)

# 5. ÿ™ŸáŸäÿ¶ÿ© Buildozer Ÿàÿ™ÿπÿØŸäŸÑ ÿßŸÑÿ•ÿπÿØÿßÿØÿßÿ™
!buildozer init
# ÿ™ÿπÿØŸäŸÑ ÿßŸÑÿ•ÿπÿØÿßÿØÿßÿ™ ŸÑÿ™ŸÜÿßÿ≥ÿ® ÿßŸÑÿ™ÿ∑ÿ®ŸäŸÇ
!sed -i 's/^title = .*/title = My Tube/' buildozer.spec
!sed -i 's/^package.name = .*/package.name = mytube/' buildozer.spec
!sed -i 's/^requirements = .*/requirements = python3,kivy==2.3.0,kivymd,yt-dlp,requests,certifi,openssl,ffpyplayer/' buildozer.spec
!sed -i 's/^android.permissions = .*/android.permissions = INTERNET,WRITE_EXTERNAL_STORAGE,READ_EXTERNAL_STORAGE/' buildozer.spec
!sed -i 's/^android.api = .*/android.api = 33/' buildozer.spec
!sed -i 's/^#android.accept_sdk_license = .*/android.accept_sdk_license = True/' buildozer.spec
!sed -i 's/^#android.ndk_api = .*/android.ndk_api = 21/' buildozer.spec

# 6. ÿ®ÿØÿ° ÿßŸÑÿ®ŸÜÿßÿ° (Ÿáÿ∞Ÿá ÿßŸÑŸÖÿ±ÿ© ÿ≥ŸäŸÜÿ¨ÿ≠ ŸÑÿ£ŸÜŸÜÿß ÿ£ÿ∂ŸÅŸÜÿß libtool)
print("üöÄ ÿ®ÿØÿ° ÿßŸÑÿ®ŸÜÿßÿ°... ŸÇÿØ Ÿäÿ≥ÿ™ÿ∫ÿ±ŸÇ 10-15 ÿØŸÇŸäŸÇÿ©.")
!yes | buildozer android debug
```
